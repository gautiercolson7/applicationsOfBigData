## Feature engineering

```python
#Drop ID
df = df.drop(['SK_ID_CURR'],axis=1)

#changement donnes categorielles en numriques
df = pd.get_dummies(df)
```

```python
df.head(10)
```

```python
X = df.loc[:,df.columns!="TARGET"]
y = df["TARGET"]

#replace Nan values with median
imputer = SimpleImputer(missing_values=np.nan,strategy='median')
X = imputer.fit_transform(X)

#Transformer (Robust : based on median)
X_robust = RobustScaler().fit_transform(X)

#Feature selection with VarianceThresholding
selector = VarianceThreshold(threshold=0.5)
X_robust = selector.fit_transform(X_robust)
#selector.get_support()

X_robust.shape
```

## XGBoost

## Splitting dataset into test dataset and training set

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
!pip install xgboost
```

```python
# Training XGBoost on the Training set
from xgboost import XGBClassifier
classifier = XGBClassifier()
classifier.fit(X_train, y_train)
```

```python
# Confusion Matrix
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

# Random Forest regression

```python
# Training the Random Forest Regression model on the whole dataset
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X, y)

# Predicting a new result
regressor.predict()

# Visualising the Random Forest Regression results (higher resolution)
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Paid')
plt.ylabel('Repaid')
plt.show()
```
